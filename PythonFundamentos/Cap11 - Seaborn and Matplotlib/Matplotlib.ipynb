## Gráfico de linha

```python
#dados
x = linspace(0,5,10)
y = x ** 2

print(x,y)
```

```python
#cria a figura
fig = plt.figure()

#define a ecala dos eixos
axes = fig.add_axes([0,0,0.8,0.8])

#cria o plot
axes.plot(x,y,'r') #esse já é o gráfico de linhas

#labels e titulo
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de linhas');
```

## subplots

```python
# divide a área de plotagem em dois subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para criar cada plot
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')

fig.tight_layout()
```

```python
#cria os subplots
fig, axes = plt.subplots(1,2, figsize = (10,4))

#cria o plot1
axes[0].plot(x,y, x, exp(x))
axes[0].set_title('Escala Padrão')

#cria o plot2
axes[1].plot(x,y,x,exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)')
```

## grid

```python
#cria os subplots
fig, axes = plt.subplots(1,2, figsize = (10,4))

#grid padrão
axes[0].plot(x,y, x, exp(x))
axes[0].set_title('Grid padrão')
axes[0].grid(True)

#grid customizado
axes[1].plot(x,y, x, exp(x))
axes[1].set_title('Grid Customizado')
axes[1].grid(color = 'darkgrey', alpha = 0.7, linestyle = "dashed", linewidth = 0.6)
```

## hist with normal generated data

```python
#dados
sample = np.random.randn(100000)

#cria figuras
fig, axes = plt.subplots(1,2,figsize = (12,4))

# plot 1
axes[0].hist(sample)
axes[0].set_title('Histograma Padrão')
axes[0].set_xlim(min(sample), max(sample))

#plot 2
axes[1].hist(sample, cumulative = True, bins = 50)
axes[1].set_title('Histograma Cumulativo')
axes[1].set_xlim(min(sample), max(sample))
```

## 3d plots

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
#Dados
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

#funcao para um mapa de cores
def ColorMap(phi_m, phi_p):
    return (+ alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

#mais dados
phi_m = np.linspace(0,2*np.pi, 100)
phi_p = np.linspace(0,2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T
```

```python
fig = plt.figure(figsize= (14,6))

#Adiciona o subplot 1 com projeção 3d
ax = fig.add_subplot(1,2,1, projection = '3d')
p = ax.plot_surface(X,Y,Z, rstride = 4, cstride = 4, linewidth = 0)

#Adiciona o subplot2 com projeção 3d
ax = fig.add_subplot(1,2,2, projection = '3d')
p = ax.plot_surface(X,Y,Z, rstride = 1, cstride = 1, cmap = cm.coolwarm, linewidth = 0, antialiased = False)

#Cria a barra de cores como legenda
cb = fig.colorbar(p,shrink = 0.5)
```
