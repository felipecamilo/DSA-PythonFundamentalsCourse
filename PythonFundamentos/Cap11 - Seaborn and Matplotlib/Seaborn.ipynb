## Usando Matplotlib, Seaborn, NumPy e Pandas na criao de grficos

```python
#Valores randmicos
np.random.seed(42)
n = 1000
perct_smokers = 0.2

#variaveis
flag_fumante = np.random.rand(n) < perct_smokers
idade = np.random.normal(40,10,n)
altura = np.random.normal(170,10,n)
peso = np.random.normal(70,10,n)#ta considerando que so independentes mas claramente no seriam

# Dataframe
dados = pd.DataFrame({'altura':altura, 'peso':peso, 'flag_fumante':flag_fumante})

#cria os dados para a varivel flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True:'Fumante',False:'No Fumante'})
```

```python
dados
```

```python
sea.set(style = "ticks")

#lmplot
sea.lmplot(x = 'altura', y = 'peso',data = dados, hue = 'flag_fumante', palette = ['tab:blue','tab:orange'], height = 7)

#labels e ttulo
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relaao linear entre Altura e peso de fumantes e no fumantes')
```
